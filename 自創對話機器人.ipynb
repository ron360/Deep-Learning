```python
import aisuite as ai
```

```python
import os
from google.colab import userdata
```

```python
# 使用 Groq api
api_key = userdata.get('Groq')
os.environ['GROQ_API_KEY']=api_key
provider = "groq"
model = "llama-3.3-70b-versatile"
```

```python
def reply(system="請用台灣習慣的中文回覆。",
          prompt="hi",
          provider="groq",
          model="llama-3.3-70b-versatile"
          ):

    client = ai.Client()

    messages = [
        {"role": "system", "content": system},
        {"role": "user", "content": prompt}
    ]


    response = client.chat.completions.create(model=f"{provider}:{model}", messages=messages)

    return response.choices[0].message.content
```

```python
# 賦予人設
system="""
請用台灣習慣的中文來寫這段 po 文：
請用企管系學生的方式思考, 也就是什麼都以框架化的方式回覆,
例如SWOT或是痛點分析,
並且經常使用抽象又模糊的詞彙。
可以適度使用列點的方式表達。
"""
```

```python
prompt = "今天咖啡灑到電腦上了!"
print(reply(system=system, prompt=prompt))
```

```python
import gradio as gr
```

```python
system="""
請用台灣習慣的中文來寫這段 po 文：
請用企管系學生的方式思考, 也就是什麼都以框架化的方式回覆,
例如SWOT或是痛點分析,
並且經常使用抽象又模糊的詞彙。
可以適度使用列點的方式表達。
"""
```

```python
provider="groq"
model="llama-3.3-70b-versatile"
```

```python
def respond(prompt):
    response = reply(system=system,
                     prompt=prompt,
                     provider = provider,
                     model = model
                    )
    return response
```

```python
# 將產生器改為符合自己主題
with gr.Blocks(title="企管系思考產生器") as demo:
    gr.Markdown("### 企管系式思考產生器")
    gr.Markdown("請輸入一則你看到的商業新聞，讓我幫你用企管系學生的方式思考，並重新詮釋！")

    with gr.Row():
        user_input = gr.Textbox(label="今天發生的事件是…", placeholder="例如：中國限制稀土出口")

    submit_btn = gr.Button("生成回覆!")
    output = gr.Textbox(label="📣 企管系專業回覆", lines=10)

    submit_btn.click(fn=respond, inputs=user_input, outputs=output)
```

```python
demo.launch(share=True, debug=True)
```
