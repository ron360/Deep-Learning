```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
```

### 1. 讀入 MNSIT 數據集

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### 2. 資料整理

### Channel

CNN 要注意一張圖有多少個 channel, 開始我們因為只有灰階, 所以只有一個 channel。因此我們要轉一下我們的資料格式:

    (28,28) --> (28, 28, 1)

```python
x_train = x_train.reshape(60000, 28, 28, 1) / 255
```

```python
x_test = x_test.reshape(10000, 28, 28, 1) / 255
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

### 3. step 1: 打造函數學習機 (CNN)

```python
model = Sequential()
```

```python
model.add(Conv2D(16, (3,3), padding='same',
                input_shape=(28,28,1),
                activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(32, (3,3), padding='same',
                activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(64, (3,3), padding='same',
                activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(128, (3,3), padding='same',   # 多一套卷積、池化層
                activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(100, activation='relu')) # 神經元數改成100
```

```python
model.add(Dense(100, activation='relu')) # 再多一層隱藏層
```

```python
model.add(Dense(10, activation='softmax'))
```

#### 看一下我們的神經網路

```python
model.summary()
```

#### 組裝

```python
model.compile(loss='categorical_crossentropy', # loss function 改用categorical_crossentropy
  optimizer=Adam(learning_rate=0.001), metrics=['accuracy']) # 使用 Adam可以更快收斂
```

### 4. step 2: fit

```python
model.fit(x_train, y_train, batch_size=300, epochs=10) # 一次300筆資料 更新改為10次
```

### Step 3. 預測

```python
result = model.predict(x_test)
result = np.argmax(result, axis=1)
```

### 看看測試資料表現如何

```python
loss, acc = model.evaluate(x_test, y_test)
```

```python
print(f'測試資料的正確率為 {acc*100:.2f}%')
```

```python
def my_predict(n):
    print('我可愛的 CNN 預測是', result[n])
    X = x_test[n].reshape(28,28)
    plt.imshow(X, cmap='Greys')
```
