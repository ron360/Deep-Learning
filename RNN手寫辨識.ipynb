```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 1. 讀入深度學習套件

```python
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import imdb
```

### 2. 讀入數據

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)
```

### 3. 資料處理

```python
x_train = sequence.pad_sequences(x_train, maxlen=200) # 改成看200字，較能正確判斷
x_test = sequence.pad_sequences(x_test, maxlen=200)
```

### 4. step 01: 打造一個函數學習機

```python
model = Sequential()
```

```python
model.add(Embedding(10000, 128))
```

```python
model.add(LSTM(256, return_sequences=True, dropout=0.2)) #將第一層改成256單元
```

```python
model.add(LSTM(128, dropout=0.2)) #設置第二層LSTM
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
             optimizer=Adam(learning_rate=0.0001), #設置Adam的learning rate 為0.0001 收斂更明顯
             metrics=['accuracy'])
```

```python
model.summary()
```

### 5. step 02: 訓練

```python
model.fit(x_train, y_train, batch_size=32, epochs=10,
         validation_data=(x_test, y_test))
```

### 換個存檔方式

這次是把 model 和訓練權重分開存, 使用上更有彈性。

```python
model_json = model.to_json()
open('imdb_model_architecture.json', 'w').write(model_json)
model.save_weights('imdb_model_weights.h5')
```
